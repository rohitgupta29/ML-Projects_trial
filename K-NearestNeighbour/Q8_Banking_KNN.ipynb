```python

  
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
df_new = df.apply(label_encoder.fit_transform)
df_new
```

```python
X = df_new[['age', 'job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week',
 'duration', 'campaign', 'pdays', 'previous', 'poutcome', 'emp_var_rate', 'cons_price_idx', 'cons_conf_idx', 'euribor3m', 'nr_employed']]
```

### KNN implementation after splitting the data

```python
#KNN implementation

X_train,X_test,y_train,y_test=train_test_split(X,yNet, test_size=0.2,random_state=5)


knn=KNeighborsClassifier(5)
knn.fit(X_train, y_train)

p=knn.predict(X_test)
```

### Accuracy Score

```python
accScore=accuracy_score(y_test, p)
print("Accuracy score:", accScore)
```
